# Generate Datasets

## 2D dataset with 2 classes

#### Create a dataset

```python
from sklearn.datasets.samples_generator import make_blobs
import pandas as pd

data, label = make_blobs(n_samples=100, centers=2, n_features=2)
df_data_2d = pd.DataFrame(dict(x=data[:,0], y=data[:,1], label=label)) # converting it into dataframe to make it easier
```

#### Plot the dataset

```python
from matplotlib import pyplot as plt

# scatter plot, dots colored by class value
colors = {0:'b', 1:'g', 2:'r', 3:'c', 4:'m', 5:'y', 6:'k'}
fig, ax = plt.subplots()
grouped = df_data_2d.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
plt.show()
```

#### Save the data to a file

```python
import pandas as pd
df_data_2d.to_csv('../data/2d_classification.csv', header=True, index=False)
```
