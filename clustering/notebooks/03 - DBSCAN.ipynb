# DBSCAN

```python
import pandas as pd

df_data = pd.read_csv('../data/2d_data_para.csv')
```

```python
df_data.head()
```

```python
data = df_data.copy()
```

```python
from matplotlib import pyplot as plt

x = data['x']
y = data['y']

plt.figure(figsize=(8,6))
plt.subplot(1,1,1)
plt.scatter(x, y, c='g', alpha=1)
plt.show()
```

### Run K-Means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4)
kmeans.fit(data)
predictions = kmeans.fit_predict(data)
```

```python
df_data['pred_label'] = predictions
```

### Find the centers of the clusters

```python
print(kmeans.cluster_centers_)
```

```python
from matplotlib import pyplot as plt
import numpy as np

colors = {0:'b', 1:'g', 2:'r', 3:'c', 4:'m', 5:'y', 6:'k'}

plt.figure(figsize=(8,6))
plt.figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')

plt.subplot(1,2,2)
unq_labels = np.unique(df_data['pred_label'])
for i in unq_labels:
    df = df_data.loc[df_data['pred_label'] == i][['x','y']]
    x = df['x']
    y = df['y']
    plt.scatter(x, y, c=colors[i], alpha=1)
    plt.title('Predictions')
    
plt.show()
```

## DBSCAN

<img src='img/dbscan.gif'>

[Image Src](https://towardsdatascience.com/the-5-clustering-algorithms-data-scientists-need-to-know-a36d136ef68)

```python
from sklearn.cluster import DBSCAN

eps = 5
min_samples = 5
clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(data)
predictions = clustering.labels_

df_data['pred_label'] = predictions
```

```python
np.unique(predictions)
```

```python
colors = {0:'b', 1:'g', 2:'r', 3:'c', 4:'m', 5:'y', 6:'k'}

plt.figure(figsize=(8,6))
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(1,1,1)

unq_labels = np.unique(df_data['pred_label'])
for i in unq_labels:
    df = df_data.loc[df_data['pred_label'] == i][['x','y']]
    x = df['x']
    y = df['y']
    plt.scatter(x, y, c=colors[i%len(colors)], alpha=1)
    
plt.show()
```

## Hyperparameters

__1. $ \epsilon $__ - Radius to be considered to find the dense regions
<br>
__2. mpts (or min_samples)__ - - Number of points to make the region marked by $ \epsilon $ dense


```python
from sklearn.cluster import DBSCAN

eps = 1
min_samples = 5
clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(data)
predictions = clustering.labels_

df_data['pred_label'] = predictions

colors = {0:'b', 1:'g', 2:'r', 3:'c', 4:'m', 5:'y', 6:'k'}

plt.figure(figsize=(8,6))
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(1,1,1)

unq_labels = np.unique(df_data['pred_label'])
print('Cluster Labels -> ', unq_labels)
for i in unq_labels:
    df = df_data.loc[df_data['pred_label'] == i][['x','y']]
    x = df['x']
    y = df['y']
    plt.scatter(x, y, c=colors[i%len(colors)], alpha=1)
    
plt.show()
```

### Another Dataset

```python
import pandas as pd
df_data = pd.read_csv('../data/2d_data_sin.csv')

from matplotlib import pyplot as plt

data = df_data.copy()
x = data['x']
y = data['y']

plt.figure(figsize=(8,6))
plt.subplot(1,1,1)
plt.scatter(x, y, c='g', alpha=1)
plt.show()
```

```python
from sklearn.cluster import DBSCAN

eps = 0.5
min_samples = 5
clustering = DBSCAN(eps=eps, min_samples=min_samples).fit(data)
predictions = clustering.labels_

df_data['pred_label'] = predictions

colors = {0:'b', 1:'g', 2:'r', 3:'c', 4:'m', 5:'y', 6:'k'}

plt.figure()
plt.figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.subplot(1,1,1)

unq_labels = np.unique(df_data['pred_label'])
print('Cluster Labels -> ', unq_labels)
for i in unq_labels:
    df = df_data.loc[df_data['pred_label'] == i][['x','y']]
    x = df['x']
    y = df['y']
    plt.scatter(x, y, c=colors[i%len(colors)], alpha=1)
    
plt.show()
```
